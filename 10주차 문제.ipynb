실습욜 매출 데이터
-product:앞뒤 공백 섞임
-price  :결축값이 2개
-quantity:결축값이 1개
sales_raw = pd.DataFrame({
    'product':[" Apple", "Banana ", "Orange","Milk", "Bread"]
    'price' :[ 1200,800,None,None,1650,2000]
    'quantity'[5,3,4,4,none,2]
})

```python
import pandas as pd


sales_raw = pd.DataFrame({
    'product': [" Apple", "Banana ", "Orange", "Milk", "Bread", "Egg"],
    'price': [1200, 800, None, None, 1650, 2000],
    'quantity': [5, 3, 4, 4, None, 2]
})


sales_df = sales_raw.copy()


sales_df['product'] = sales_df['product'].str.strip()


sales_df['price'] = sales_df['price'].fillna(
    sales_df['price'].mean()
)

sales_df['quantity'] = sales_df['quantity'].fillna(
    sales_df['quantity'].mean()
)


sales_df
```
