복습 문제 - 함수

```python
def calculate_average(numbers):
    total = 0
    for n in numbers:
        total += n
    return total / len(numbers)

my_score = [80, 90, 70, 100]
avg = calculate_average(my_score)
print("평균:", avg)
```

복습 문제 - 종합

Q)
3명의 학생 이름과 점수가 있고 점수에 따라 등급을 매기는 함수를 생성하고 로그를 출력합니다.

EX)
- A 학생은 A등급입니다.
- B 학생은 C등급입니다.
- C 학생은 B등급입니다.

```python
students = ["A", "B", "C"]
students_scores = [77, 88, 99]

def calculate_score(name, num):
    if num < 70:
        print(f"{name}학생은 F등급입니다.")
    elif num < 80:
        print(f"{name}학생은 C등급입니다.")
    elif num < 90:
        print(f"{name}학생은 B등급입니다.")
    else:
        print(f"{name}학생은 A등급입니다.")

for count in range(len(students)):
    calculate_score(students[count], students_scores[count])
```

# 머신 러닝

> 정의: 사람이 규칙을 직접 생성하는 대신,  
> 데이터를 보고 컴퓨터가 스스로 규칙을 찾아내는 방법

일반 프로그래밍 vs 머신러닝

| 구분             | 일반 프로그래밍             | 머신러닝                      |
| ---             | ---                       | ---                          |
| 규칙을 누가 만드나 | 사람이 직접 만든다          | 컴퓨터가 찾는다                |
| 필요한 것         | 규칙 (If문 등)            | 데이터 + 정답                  |
| 예시             | if 점수 >= 60: 합격       | 과거 데이터를 참조, 기준을 찾게함 |
| 장점             | 규칙이 적거나, 명확할때 좋음 | 규칙이 복잡할 때 좋음           |

# 모델 학습과 예측 실습

## 모델이란?

> 정의: 데이터를 보고 규칙을 학습한 뒤.
> 새로운 데이터가 들어오면 결과를 예측하는 프로그램

쉬운 비유:
- 모델 = 학생
- 학생(모델)에게 문제와 답, 개념(데이터)을 주고 공부를 시킵니다.
- 공부가 끝나면, 새로운 문제(데이터)를 풀게 합니다.

## 학습(fit)이란?

> 정의: 모델이 입력값(x)과 정답(y)을 보고
둘 사이의 패턴이나 규칙을 찾는 과정

- `model.fit(x, y)`을 사용하면 학습이 시작됩니다.
- x(입력값) 와 y(정답)을 함께 넣어줘야 합니다.
- 모델은 학습이 끝나고 규칙을 기억하고 있게 됩니다.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "study_hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "attendance": [50, 55, 60, 80, 60, 76, 5, 80],
    "pass_fail": [0, 0, 0, 1, 0, 1, 1, 1,]
}

df = pd.DataFrame(data)

# 입력값과 정답 분리
x = df[["study_hours", "attendance"]]
y = df["pass_fail"]

print("데이터 준비 완료")
print(f"학생 수: {len(df)}명") # 학생 수: %%명
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(x, y)
print("모델 학습 완료")
```

```python
# 새 학생 1명을 예측해보기
# 주당 4시간 공부. 출석률을 80%
new_student = pd.DataFrame([[4, 80]], columns=["study_hours", "attendance"])
prediction = model.predict(new_student)
print("공부 4시간, 출석 80%인 학생:", prediction)

if prediction[0] == 1:
    print("모델 판단: 합격")
else:
    print("모델 판단: 불합격")
```

```python
# 여러 학생 예측하기

new_student = pd.DataFrame([
    [1, 50],
    [3, 65],
    [5, 80],
    [7, 95],
], columns=["study_hours", "attendance"])

results = model.predict(new_student)

# 문제:
# 각 학생의 로그를 출력하세요.
# 예시:
# 공부 ?시간, 출석 ?% 인 학생은 합격 입니다.
# 공부 ?시간, 출석 ?% 인 학생은 불합격 입니다.
# 공부 ?시간, 출석 ?% 인 학생은 합격 입니다.
# 공부 ?시간, 출석 ?% 인 학생은 불합격 입니다.

for i in range(len(new_student)):
    study = new_student.iloc[i]["study_hours"]
    attend = new_student.iloc[i]["attendance"]
    
    if results[i] == 1:
        print(f"공부 {study}시간, 출석 {attend}% 인 학생은 합격 입니다.")
    else:
        print(f"공부 {study}시간, 출석 {attend}% 인 학생은 불합격 입니다.")
```
